```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
import numpy as np

# Load DEM GeoTIFF
dem_file = "h12324_utm.tif"
with rasterio.open(dem_file) as dem:
    elevation_data = dem.read(1)
    show(dem, title="River Basin Elevation Map")

low_areas = np.where(elevation_data < 150, 1, 0)

# Visualize low elevation areas
plt.imshow(low_areas, cmap="Blues", alpha=0.6)
plt.title("Flood Risk Zones Based on Elevation (<150m)")
plt.colorbar(label="Flood Risk")
plt.show()
```
